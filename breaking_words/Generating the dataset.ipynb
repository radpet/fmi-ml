```python
def download_and_extract(url, path):
    request = requests.get(url)
    with ZipFile(BytesIO(request.content), "r") as file:
         file.extractall(path)
        
for url in urls:
    download_and_extract(url,"./data/")
```

```python
import pandas as pd
import os
```

```python
def create_dataset():
    data = []
    for filename in os.listdir("./data"):
        file = open("./data/{}".format(filename))
        data.append( (filename,file.read().strip()) )
        file.close()
        os.remove("./data/{}".format(filename))
    return pd.DataFrame(data=data,index=range(len(data)),columns=["title","text"])

data = create_dataset()
data
```

Сега ще използвам текста, за да създам корпус от думи

```python
import nltk
nltk.download("punkt")
```

```python
def lower(str):
    return str.lower()
    
data["tokenized_text"] = data["text"].apply(lower).apply(nltk.word_tokenize)
```

```python
#pip3 install alphabet-detector

from alphabet_detector import AlphabetDetector
ad = AlphabetDetector()

def not_num(str):
    try:
        float(str)
        return False
    except:
        return True
    
def no_bad_char(str):
    return not ("." in str or "_" in str or '"' in str or "„" in str or "…" in str or "*" in str or "/" in str or "," in str) and ad.is_cyrillic(str)

corpus = set()
cleaned = data["tokenized_text"].apply(lambda tokens: set(filter(lambda token: not_num(token) and no_bad_char(token), tokens) ))
for c in [x for x in cleaned]:
    corpus = corpus | c

print(len(corpus))
corpus
```

```python
sorted_corpus = [word for word in corpus]
sorted_corpus.sort();

file = open("./data/corpus","w")

for word in sorted_corpus:
    file.write(word)
    file.write("\n")
file.close()
```

```python
from nltk import ngrams

def get_n_grams(str ,n):
    return ngrams([ w.lower().strip() for w in str.split(" ") ], n)
```
